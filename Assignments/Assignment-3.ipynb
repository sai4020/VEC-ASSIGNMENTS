```python
d = data.drop(columns=['Id'],axis=1)
d
```

Preprocessing of data

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
d['Species']=lb.fit_transform(d['Species'])
d
```

Data Visualization

```python
sns.displot(d['Species'])
```

Separating the Data

```python
x = d.drop('Species',axis=1)
x
```

```python
y=d['Species']
y
```

Scaling of Data

```python
from sklearn.preprocessing import MinMaxScaler
Scale = MinMaxScaler()
S = pd.DataFrame(Scale.fit_transform(x),columns=x.columns)
S
```

Splitting of Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(S, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
p=model.predict(x_test)
p
```

```python
y_test
```

Evaluating Model Performance

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,p)
```
